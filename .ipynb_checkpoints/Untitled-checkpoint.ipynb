```python
import scipy.stats
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(123)
```

```python
mean = 2.0
std = 3.0
for N in [3, 100, 10000]:
    x = scipy.stats.norm.rvs(loc=mean, scale=std, size=N)
    print('N = %d  mean = %f  std = %e' % (N, x.mean(), x.std(ddof=1)))
    plt.figure()
    plt.hist(x, bins=20)
    plt.title('$N = %i$' % (N))
```
